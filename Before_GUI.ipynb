```python
import pandas as pd
import numpy as np

# Create an empty DataFrame to store the label counts for each input file
output_df = pd.DataFrame(columns=['file_name', 'positive', 'negative', 'neutral'])

while True:
    file_name = input("Enter the name of a file (or 'exit' to quit): ")
    if file_name == 'exit':
        break
    else:
        dataset = pd.read_csv(f'/content/Output_Files/test_data/{file_name}.csv')
        # process the dataset here
        dataset.columns.values[0] = "text"
        dataset['label'] =0
        dataset['text'] = dataset['text'].apply(handle_emojis, args=[emojis])
        dataset['text'] = dataset['text'].apply(lambda x: data_cleaning(x))
        dataset['text'] = dataset['text'].apply(lambda x: remove_diacritics(x))
        dataset = dataset.dropna()
        dataset = dataset.drop_duplicates()

        result, model_outputs, wrong_predictions = model.eval_model(dataset)

        # Get the index of the maximum value for each row
        max_index_per_row = np.argmax(model_outputs, axis=1)

        # Apply the decode function to convert the maximum value indices to labels
        decoded_labels = [decode(x) for x in max_index_per_row]

        # Create a new DataFrame with the text and decoded labels
        data = [[text, label] for text, label in zip(dataset['text'], decoded_labels)]
        new_df = pd.DataFrame(data, columns=['text', 'label'])
        print(new_df)

        # Count the number of occurrences of each label
        label_counts = new_df['label'].value_counts()

        # Add the label counts to the output DataFrame
        output_df = output_df.append({'file_name': file_name, 'positive': label_counts[0], 'negative': label_counts[2], 'neutral': label_counts[1]}, ignore_index=True)

# Print the resulting DataFrame

output_df
```
